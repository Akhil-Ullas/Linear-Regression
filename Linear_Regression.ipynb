# **1. Linear Regression**

* Linear Regression is used for estimating the relationships among variables.
* The goal of the algorithm is to find the best linear equation that can
predict the value of the dependent variable based on the independent
variables
* Used for regression data

1.  Simple   -> y=mx+c(Straigt Line),only 1 x values
2.  Multiple -> y=m1x1+m2x2+.....mnxn+b,more than 1 x values
3.  Polynomial

# **Salary Data**

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

#machine learning Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/drive/MyDrive/Camerin DSML/Salary.csv')
df.head(15)
```

**Step 1 : Preprocessing**

```python
#checking Shape
df.shape     #35 rows and 2 columns
```

```python
#checking null values
df.isna().sum() #no null values
```

```python
#checking number od datatypes
df.dtypes    #'YearsExperience' -> float
             #'Salar'           -> int
```

**Step 2 : Model Selection**

```python
x=df.drop(columns='Salary')  #x->YearsExperience
y=df['Salary']
x.head()
```

```python
y.head()   #y-> Salary
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
```

**Step 3 : Training & Testing**

```python
model=LinearRegression()
model.fit(x_train,y_train)

pred_y=model.predict(x_test)
pred_y
```

**Step 4 : Model Evaluation**

```python
a=model.score(x_test,y_test)
print(f"Accuracy of the model = {np.round((a*100),2)}%")
```

**step 5 : Model Deployment**

```python
new_data=pd.DataFrame([5.5],columns=x.columns)
prediction=model.predict(new_data)
print(f"Salary = {np.round(prediction,2)}Rs")
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/drive/MyDrive/Camerin DSML/student_exam_scores.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.dtypes
```

```python
from sklearn import preprocessing
label_encode=preprocessing.LabelEncoder()
df['student_id']=label_encode.fit_transform(df['student_id'])
df['student_id'].unique()
```

```python
df.dtypes
```

```python
x=df.drop(columns='exam_score')
x.head()
```

```python
y=df['exam_score']
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)

pred_y=model.predict(x_test)
pred_y
```

```python
a=model.score(x_test,y_test)
print(f"Accuracy of the model = {np.round((a*100),2)}%")
```

```python
# Create a DataFrame to hold actual and predicted values
comparison_df=pd.DataFrame({
    'Actual':y_test.values,
    'Predicted':pred_y
})

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(comparison_df['Actual'],label='Actual',marker='o')
plt.plot(comparison_df['Predicted'],label='Predicted',marker='x')
plt.title('Actual vs Predicted Exam Scores')
plt.xlabel('Sample Index')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()

```

```python
df.head()
x=df.drop(columns='exam_score')
a=pd.DataFrame([[1,9,8,80,79]],columns=x.columns)
prediction=model.predict(a)
print(f'Predicted exam score:{np.round(prediction,2)}')
```
